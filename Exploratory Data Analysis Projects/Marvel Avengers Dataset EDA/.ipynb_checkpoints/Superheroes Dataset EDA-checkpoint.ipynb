```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
superheroes = pd.read_csv('charcters_stats.csv')
```

```python
superheroes.head()
```

```python
superheroes.shape
```

```python
superheroes['Alignment'].value_counts()
```

```python
# Get the good superheroes
```

```python
good = superheroes[superheroes['Alignment'] == 'good']
good.head(10)
```

```python
# Get the details of the fastest superheroes
```

```python
good.sort_values(by = ['Speed'], ascending = False).head()
```

```python
# Get the details of the most powerful superheroes
```

```python
good.sort_values(by = ['Power'], ascending = False).head()
```

```python
# Get the details of the most Deadly superheroes (Highest Total)
```

```python
good.sort_values(by = ['Total'], ascending = False).head()
```

```python
# Superheroes with 100 power
```

```python
good_max_power_full = good[good['Power'] == 100]
```

```python
good_max_power_full.head()
```

```python
good_max_power_full.sort_values(by='Total', ascending = False, inplace = True)
```

```python
good_max_power_full.head(10)
```

```python
# Create a bar chart for most powerful superheroes
```

```python
plt.figure(figsize = (8, 5))
plt.bar(list(good_max_power_full['Name'])[0:5], list(good_max_power_full['Total'])[0:5], color = 'orange')
plt.title("Most overall powerful superheroes")
plt.xlabel("name of the Superhero")
plt.ylabel("Total Power")
plt.show()
```
