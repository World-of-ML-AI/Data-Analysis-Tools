```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.shape
```

```python
fifa.head()
```

```python
# Print column names
for col in fifa.columns:
    print(col)
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize = (5, 5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]), color = 'orange')
plt.title("Countries with most number of players")
plt.xlabel("Countries")
plt.ylabel("Number of players")
plt.show()
```

```python
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary.sort_values(by=['wage_eur'], ascending = False, inplace = True)
```

```python
player_salary.head()
```

```python
plt.figure(figsize = (8, 5))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color = ['maroon', 'red', 'yellow', 'orange', 'pink'])
plt.title("Countries with most number of players")
plt.xlabel("Countries")
plt.ylabel("Number of players")
plt.show()
```
