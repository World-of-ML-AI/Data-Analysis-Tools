### Line Plot in Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri)
plt.title("Line Plot of fmri dataset")
plt.show()
```

**Grouping data with hue**

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri, hue = "event")
plt.title("Line Plot of fmri dataset")
plt.show()
```

**Adding Styles to our lines**

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri, style = "event", hue = "event")
plt.title("Line Plot of fmri dataset")
plt.show()
```

**Adding markers**

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri, style = "event", hue = "event", markers = True)
plt.title("Line Plot of fmri dataset")
plt.show()
```

## Bar Graphs in Seaborn

```python
import pandas as pd
```

```python
sns.set(style = "whitegrid")
```

```python
pokemon = pd.read_csv("Pokemon.csv", encoding='latin1')
pokemon.head()
```

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Total", data=pokemon)
plt.show()
```

```python
sns.barplot(x="Legendary", y="Total", hue="Stage", data=pokemon)
plt.show()
```

**Palette Attribute**

***1. Blues_d palette***

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon, palette="Blues_d")
plt.show()
```

***2. Rocket palette***

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon, palette="rocket")
plt.show()
```

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon, palette="vlag")
plt.show()
```

**Using the Color attribute**

```python
sns.barplot(x="Legendary", y="Speed", data=pokemon, color="orange")
plt.show()
```

## Seaborn Scatter Plot

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris)
plt.show()
```

**Changing Aesthetics**

```python
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris, hue = 'species', style = "species")
plt.show()
```

**Determine colors with respect to the numerical columns instead of categorical colums**

```python
sns.scatterplot(x = "sepal_length", y = "petal_length", data = iris, hue = 'petal_length', style="species")
plt.show()
```

## Seaborn histograms

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.histplot(diamonds['price'])
plt.show()
```

**Hide the Histogram and show the frequency curve**

```python
sns.distplot(diamonds['price'], hist =False)
plt.show()
```

**Change color of the columns in the histogram**

```python
sns.distplot(diamonds['price'], color = "green")
plt.show()
```

**Show only the histogram and not the frequency curve**

```python
sns.distplot(diamonds['price'], color = 'orange', bins = 10, kde = False)
plt.show()
```

**Vertical Mapping**

```python
sns.distplot(diamonds['price'], color='red', vertical = True)
plt.show()
```

## Joint Plot (Combination of Scatter Plot and Histogram)

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris)
plt.show()
```

**Change color of the jointplot**

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, color='olive')
plt.show()
```

**Let us also add a Regression line to our scatter plot and histograms**

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, color = 'orange', kind = 'reg')
plt.show()
```

## Box Plot

**Used to understand how a categorical value changes with a numerical value**

```python
churn = pd.read_csv('churn.csv')
```

```python
churn.head()
```

```python
churn.shape
```

```python
sns.boxplot(x='Churn', y = 'tenure', data = churn)
plt.show()
```

**Box plot between the Internet Services column and the Monthly Charges Column**

```python
sns.boxplot(x='InternetService', y = 'MonthlyCharges', data = churn)
plt.show()
```

**Box plot between the Contract column and the Tenure Column**

**Using a new Palette**

```python
sns.boxplot(x='Contract', y = 'tenure', data = churn, palette = 'Set1')
plt.show()
```

**Change the thickness of the boundary lines**

```python
sns.boxplot(x='Contract', y = 'tenure', data = churn, linewidth = 3)
plt.show()
```

**Change the order of the columns**

```python
sns.boxplot(x='Contract', y = 'tenure', data = churn, palette = 'Set1', order = ["Two year", "Month-to-month", "One year"])
plt.show()
```

**Hue attribute**

```python
sns.boxplot(x='Contract', y = 'tenure', data = churn, hue = "PaymentMethod")
plt.show()
```
